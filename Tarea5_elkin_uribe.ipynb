<span style="color:green"><justify>1. Cree dos archivos de tipo csv. En el primero debe colocar la información de 100 personas (clientes): id, cedula, nombre, apellido, número de cuenta.  En el segundo debe colocar transacciones de cada cliente. Por ejemplo (id= 3456, fecha = 29/01/2023, valor = -50_000). En total  mil transacciones. </justify></span>

```python
import csv
import random
```

```python
# Función para generar una cédula aleatoria
def generar_cedula():
    cedula = ''.join(random.choices('0123456789', k=10))
    return cedula

# Generar el archivo CSV de clientes
with open('clientes.csv', mode='w', newline='') as clientes_file:
    writer = csv.writer(clientes_file)
    writer.writerow(['id', 'cedula', 'nombre', 'apellido', 'numero_cuenta'])

    for i in range(100):
        id_cliente = i + 1
        cedula_cliente = generar_cedula()
        nombre_cliente = f'Cliente_{id_cliente}'
        apellido_cliente = f'Apellido_{id_cliente}'
        numero_cuenta = f'CTA{i + 1}'

        writer.writerow([id_cliente, cedula_cliente, nombre_cliente, apellido_cliente, numero_cuenta])

# Generar el archivo CSV de transacciones
with open('transacciones.csv', mode='w', newline='') as transacciones_file:
    writer = csv.writer(transacciones_file)
    writer.writerow(['id', 'fecha', 'valor'])

    for i in range(1000):
        id_cliente = random.randint(1, 100)
        fecha = f'{random.randint(1, 31)}/{random.randint(1, 12)}/2023'
        valor = random.randint(-100000, 100000)

        writer.writerow([id_cliente, fecha, valor])

print("Archivos CSV generados exitosamente.")
```

<span style="color:green"><justify>3. Construya una función  que haga lo mismo con el segundo archivo.</justify></span>

```python
import pandas as pd
```

```python
def cargar_transacciones_csv(file_path):
    df_transacciones = pd.read_csv(file_path)
    return df_transacciones

# Ejemplo de uso de la función
transacciones_file_path = 'transacciones.csv'
df_transacciones = cargar_transacciones_csv(transacciones_file_path)

# Imprimir el DataFrame cargado
print(df_transacciones.head())
```

<span style="color:green"><justify>4.  Construya una función que reciba los dos objetos de Pandas y cree un diccionario, en donde hay solamente un registro por cada cliente. Por ejemplo ( esto es solamente un ejemplo)  {id= 2345, nombre='Luis Vazquez', cedula=20456912, transacciones=[ {fecha: 0/12/2021, valor: 45_000}, {fecha: 15/02/2023; valor: -70_000}, ...]}</justify></span>

```python
def crear_diccionario_clientes_transacciones(df_clientes, df_transacciones):
    # Crear diccionario vacío
    diccionario_clientes = {}

    # Iterar sobre cada fila del DataFrame de clientes
    for index, row in df_clientes.iterrows():
        # Obtener los datos del cliente
        id_cliente = row['id']
        nombre_cliente = row['nombre']
        cedula_cliente = row['cedula']

        # Filtrar las transacciones del cliente en el DataFrame de transacciones
        transacciones_cliente = df_transacciones[df_transacciones['id'] == id_cliente][['fecha', 'valor']].to_dict(orient='records')

        # Agregar los datos del cliente y sus transacciones al diccionario
        diccionario_clientes[id_cliente] = {
            'id': id_cliente,
            'nombre': nombre_cliente,
            'cedula': cedula_cliente,
            'transacciones': transacciones_cliente
        }

    return diccionario_clientes

# Ejemplo de uso de la función
diccionario_clientes_transacciones = crear_diccionario_clientes_transacciones(df_clientes, df_transacciones)

# Imprimir el diccionario
for cliente_id, cliente_data in diccionario_clientes_transacciones.items():
    print(f"Cliente ID: {cliente_id}")
    print(f"Nombre: {cliente_data['nombre']}")
    print(f"Cédula: {cliente_data['cedula']}")
    print("Transacciones:")
    for transaccion in cliente_data['transacciones']:
        print(f"Fecha: {transaccion['fecha']}, Valor: {transaccion['valor']}")
    print()
```

<span style="color:green"><justify>5. Convierta el diccionario en un objeto json.</justify></span>

```python
import json

# Convertir el diccionario en un objeto JSON
json_objeto = json.dumps(diccionario_clientes_transacciones)

# Imprimir el objeto JSON
print(json_objeto)
```

<span style="color:green"><justify>6. Guarde el objeto json como un archivo json.</justify></span>

```python
# Ruta del archivo JSON
archivo_json = 'clientes_transacciones.json'

# Guardar el objeto JSON en un archivo
with open(archivo_json, 'w') as file:
    json.dump(diccionario_clientes_transacciones, file)

print(f"El objeto JSON ha sido guardado en el archivo '{archivo_json}'")
```
